**Señal de linea basal 1'**

Condiciones:



1.   Ojos tapados, evitando movimiento facial
2.   Oidos tapados, evitando contacto con cualquier superficie

```
import matplotlib.pyplot as plt
import numpy as np

#Antes de realizar el ploteo correspondiente, es necesario tener las siguientes consideraciones:

#La frecuencia de muestreo del bitalino que se escogio fue de 100 Hz.
#Asimismo, la funcion de tranferencia utilizada para obtener la gráfica fue:
# Voltaje = (((ADC / 1024) - 0.5) * 3.3) / 40000
# El cual cumples con las especifiaciones BITalino

#-------PLOTEO DE LA SEÑAL 1
array= np.genfromtxt("/content/1_estado_basal_1min.txt",delimiter="	",skip_header = 2) #leemos el archivo txt

M = array[:,-2].shape[0] #longitud de la señal
valores=array[:,-2]

cantidad=np.size(valores);
tiempo=np.arange(0,cantidad)
tiempo=tiempo/100

valores=(((valores/1024)-0.5)*3.3)/(40000)
valores=valores*1000000
plt.figure(figsize=[12,4])
plt.plot(tiempo, valores)
plt.ylabel('EEG Signal (uV)')
plt.xlabel('Time (s)');
plt.title('Estado basal - EEG')
plt.grid(True)
```
